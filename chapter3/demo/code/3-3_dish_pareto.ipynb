```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取餐饮菜品盈利数据 '../data/catering_dish_profit.xls'
profit_data = pd.read_excel('../data/catering_dish_profit.xls')
profit_data

# 复制‘盈利’列
data = profit_data['盈利'].copy()
#data, type(data)

# 对数据排序
data.sort_values(ascending=False)

# 设置中文标签显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 获取图像对象
plt.figure()
# 设置 纵轴 y 轴为 '盈利（元）'
plt.ylabel(u'盈利（元）')
# 绘制条形图 bar
data.plot(kind='bar')

# 求数据累积和占总和比率，记为 
p = data.cumsum()
p = p / data.sum()
#  绘制数据 cr，颜色 属性为红色，style属性 ‘-o’, 线宽 2，
p.plot(color='r', secondary_y=True, style='-o', linewidth=2)

# 注解 p 对应的图像，并指定箭头样式。
plt.annotate(s=format(p[6], '.4%'),
            xy=(6, p[6]),
            xytext=(6,p[6] * 0.9),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.9'))

# 设置y轴标签为"盈利（比率）
plt.ylabel(u'盈利（比率）')

# 显示图像
plt.show()
```
