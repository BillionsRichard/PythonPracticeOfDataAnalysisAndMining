```python
import pandas as pd
import numpy as np
```

```python
datafile = "../data/air_data.csv"
resultfile = "../tmp/explore.xls"
```

```python
data = pd.read_csv(datafile, encoding='utf8')
```

```python
data
```

```python
explore = data.describe(percentiles=[], include='all').T
```

```python
explore
```

```python
explore['null'] = len(data) - explore['count'] # 给DataFrame 新增 1 列数据，列名称：null
```

```python
explore
```

```python
explore = explore[['null', 'max', 'min']] # DF 中选取若干列的方法。
```

```python
explore
```

```python
explore.columns
```

```python
explore.columns = [u'空值数', u'最大值', u'最小值'] # 列名重命名.
```

```python
explore
```

```python
explore.to_excel(resultfile)
```
