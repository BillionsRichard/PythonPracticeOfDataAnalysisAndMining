```python
import pandas as pd
```

```python
infile = "../data/sales_data.xls"
```

```python
data = pd.read_excel(infile, index_col="序号")
```

```python
data
```

```python
data[data=='好'] = 1
data[data=='是'] = 1
data[data=='高'] = 1
data[data!=1] = 0
```

```python
data
```

```python
x = data.iloc[:, :3]
y = data.iloc[:, 3].values.astype(int)
```

```python
type(x)
```

```python
X = x.values.astype(int)
```

```python
X, X.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers.core import Dense, Activation
```

```python
model = Sequential()
```

```python
model.add(Dense(input_dim=3, output_dim=10))
```

```python
model.add(Activation('relu'))
```

```python
model.add(Dense(input_dim=10, output_dim=1))
```

```python
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(X, y, epochs=1000, batch_size=10)
```

```python
y_pred = model.predict_classes(X, verbose=1).reshape(len(y))
```
