```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
datafile = "../data/data.xls"
processedfile = "../tmp/data_processed.xls"
```

```python
typelabel = {u'肝气郁结证型系数': 'A',
            u'热毒蕴结证型系数':'B',
            u'冲任失调证型系数':'C',
            u'气血两虚证型系数':'D',
            u'脾胃虚弱证型系数':'E',
            u'肝肾阴虚证型系数':'F'}
```

```python
k = 4
```

```python
data = pd.read_excel(datafile)
```

```python
data
```

```python
keys = list(typelabel.keys())
```

```python
keys
```

```python
result = pd.DataFrame()
```

```python
for i in range(len(keys)):
    print('正在进行{}的聚类'.format(keys[i]))
    kmodel = KMeans(n_clusters=4, n_jobs=4)
    kmodel.fit(data[[keys[i]]].values)
    
    r1 = pd.DataFrame(kmodel.cluster_centers_, columns=[typelabel[keys[i]]])
    r2 = pd.Series(kmodel.labels_).value_counts()
    r2 = pd.DataFrame(r2, columns=[typelabel[keys[i]]+'n'])
    r = pd.concat([r1, r2], axis=1).sort_values(typelabel[keys[i]])
    r.index=[1, 2, 3, 4]
    c = r[typelabel[keys[i]]]
    r[typelabel[keys[i]]] = c.rolling(window=2).mean()
    r[typelabel[keys[i]]][1] = 0.0
    result = result.append(r.T)
```
